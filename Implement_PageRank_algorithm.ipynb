```python
import pandas as pd
import numpy as np
```

```python
# Initilize constant N --> Number if webpages
N = 4
```

```python
# Initilize constant d --> damping parameter
d = 0.7
```

```python
# Assuming that all of the page ranks are 1 in the beginning
pgRankA = 1.0
pgRankB = 1.0
pgRankC = 1.0
pgRankD = 1.0
```

```python
num = 5
for _ in range(num):
  pgRankA = (1 - 0.7)/4 + 0.7 * np.sum([1,pgRankC/2])
  print("Updated iteration of A page rank:")
  print(pgRankA,"\n")

  pgRankB = (1 - 0.7)/4 + 0.7 * np.sum([2,pgRankD/1])
  print("Updated iteration of B page rank:")
  print(pgRankB,"\n")

  pgRankC = (1 - 0.7)/4 + 0.7 * np.sum([2,pgRankD * pgRankA * pgRankB/3])
  print("Updated iteration of C page rank:")
  print(pgRankC,"\n")

  pgRankD = (1 - 0.7)/4 + 0.7 * np.sum([2,pgRankC/1])
  print("Updated iteration of D page rank:")
  print(pgRankD,"\n\n\n\n")
```
